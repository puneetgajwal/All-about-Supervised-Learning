```python
# Importing Libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import  cross_val_score
from collections import Counter
from sklearn.utils import shuffle
from sklearn.preprocessing import StandardScaler
```

```python
df=shuffle(df)
# Seperating labels
y=df['Class']
x=df.drop('Class',axis=1)
```

```python
x.head()
```

```python
# Standardizing Data
# Initializing satndard scalar
scaler = StandardScaler()
```

```python
scaled_x = scaler.fit_transform(x)
```

```python
scaled_x
```

```python
# Splitting data into test and train
x_train,x_test,y_train,y_test=train_test_split(scaled_x,y,test_size=0.25)
# First split with train size=75%, test size=25%
#x_test and y_test will be used in prediction
```

```python
x_train.shape
# 223 train points
```

```python
x_test.shape
# 75 points for testing
```

<p style="color:DodgerBlue;"> We will be using cross validation to find most optimal K for our data. We need to allocate data for CV as well. So from training data, we will split CV data from it </p>

```python
# splitting data into cross validation data from x_train
x_train_knn,x_cv,y_train_knn,y_cv=train_test_split(x_train,y_train,test_size=0.2)
# Second split x_cv ~45 points.
```

```python
# Prediction on iris_cv for optimum k
k=[]
accuracy=[]
for i in range(1,40,2):
    knn=KNeighborsClassifier(n_neighbors=i)
    #fitting model with k=1,2,5....39 on x_train_knn
    knn.fit(x_train_knn,y_train_knn)
    #predicting on iris_cv
    pred=knn.predict(x_cv)
    k.append(i)
    accuracy.append(acc)
    #Evaluationg Accuracy
    acc=accuracy_score(y_cv,pred,normalize=True)*float(100)
    print("\n CV k,accuracy",(i,acc))
# For K=11 we are getting highest accuracy for train data. Also for K=3,5,9 are also worth looking into
```

```python
plt.plot(k,accuracy)
```

<p style="color:DodgerBlue;"> I'll try to fit model for 3,5,9 and 11 NN on larger data without CV so that training can be done on larger part. Now we will test it on actual test Data. </p>

```python
n_list=[3,5,9,11]
final_k=[]
final_acc=[]
for k in n_list:
    knn=KNeighborsClassifier(n_neighbors=k)    
    knn.fit(x_train,y_train)
    pred=knn.predict(x_test)
    acc=accuracy_score(y_test,pred,normalize=True)*float(100)
    final_k.append(k)
    final_acc.append(acc)
    print("\n K,accuracy for shortlisted neighbors is ", (k,acc))
```

```python
plt.plot(final_k,final_acc)
```

<p style="color:DodgerBlue;"> For K=5, we are getting maximum accuracy i.e ~89.7 on actual test data. This looks like the optimal k. We will see other metrcis as well for this </p>

```python
#import confusion_matrix
from sklearn.metrics import confusion_matrix
#let us get the predictions using the classifier we had fit above
knn=KNeighborsClassifier(n_neighbors=5)    
knn.fit(x_train,y_train)
pred=knn.predict(x_test)
confusion_matrix(y_test,pred)
pd.crosstab(y_test, pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

 <p style="color:DodgerBlue;"> We can see that our model has wrongly 3 predicted Tye H points as Normal, 4 Normal points as type H and 1 Type S point as Normal. So total mis-classified points are 8</p>

```python
# Classification_report
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```

 <p style="color:DodgerBlue;"> <B> Problem That we are facing : </B> Here we have an imbalanced dataset with only 480 variables classified as LoanOnCard=1 and 0 for 4500 variables. We cannot do a random sample as well in this case. <br> <br>
    <B> Strategy we'll be using : </B> We will be creating Synthetic points for undersampled class so that our model could train properly on it . We will be using  <b>Synthetic Minority Over-sampling Technique (SMOTE) </B>.
</p><br>

```python
# Seperating labels
LoanOnCard=df['LoanOnCard']
x=df.drop('LoanOnCard',axis=1)
```

```python
# scatter plot of examples by class label
pd.value_counts(df['LoanOnCard']).plot.bar()
plt.title('LoanOnCard class histogram')
plt.xlabel('Class')
plt.ylabel('Frequency')
df['LoanOnCard'].value_counts()
```

```python
x = StandardScaler().fit_transform(x)
```

```python
from imblearn.over_sampling import SMOTE
X_train, X_test, y_train, y_test = train_test_split(x, LoanOnCard, test_size=0.3, random_state=0)

print("Number Observations in X_train dataset: ", X_train.shape)
print("Number Observations in y_train dataset: ", y_train.shape)
print("Number Observations in X_test dataset: ", X_test.shape)
print("Number Observations in y_test dataset: ", y_test.shape)
```

```python
print("Before OverSampling, counts of label '0': {} \n".format(sum(y_train==0)))

sm = SMOTE(random_state=2)
X_train_res, y_train_res = sm.fit_sample(X_train, y_train.ravel())

print('After OverSampling, the shape of train_X: {}'.format(X_train_res.shape))
print('After OverSampling, the shape of train_y: {} \n'.format(y_train_res.shape))

print("After OverSampling, counts of label '1': {}".format(sum(y_train_res==1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train_res==0)))
```

```python
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_train_res, y_train_res, test_size=0.3, random_state=0)

print("Number Observations in X_train_s i.e after SMOTE dataset: ", X_train_s.shape)
print("Number Observations in y_train_s i.e after SMOTE dataset: ", y_train_s.shape)
print("Number Observations in X_test_s i.e after SMOTE dataset: ", X_test_s.shape)
print("Number Observations in y_test_s i.e after SMOTE dataset: ", y_test_s.shape)
```

```python
pd.value_counts(y_train_s).plot.bar()
plt.title('LoanOnCard class histogram after SMOTE in train data')
plt.xlabel('Class')
plt.ylabel('Frequency')
```

```python
pd.value_counts(y_test_s).plot.bar()
plt.title('LoanOnCard class histogram after SMOTE in Test data')
plt.xlabel('Class')
plt.ylabel('Frequency')
```
